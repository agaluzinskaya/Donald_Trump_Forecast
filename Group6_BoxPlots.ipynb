```python
import pandas 
import matplotlib.pyplot as plt

data = pandas.read_csv('DonaldTrump_SentimentAnalysis.csv', sep=',')
ax = data.groupby('election_status')
ax.boxplot(return_type='axes',column=['perc_likes','perc_love','perc_hahas','perc_wows','perc_sads','perc_angrys'], 
           rot=90, figsize=(8, 6), medianprops={'linestyle': '--', 'linewidth': 1})

plt.show()
```
